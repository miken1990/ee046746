```python
import json

with open('my_data/imagenet_classes_json/imagenet_class_index.json', 'r') as f:
    labels = json.load(f)

labels = {int(key): value[1] for key, value in labels.items()}
size = 244
```

```python
# show class of each bird image with our image
from part2.pre_process import prepare_image, predict
from PIL import Image
tr_pipeline = transforms.Compose([transforms.Resize(size),
                                         transforms.ToTensor(),
                                         transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                              std=[0.229, 0.224, 0.225])])

bird_dir_path = './birds'
# plot given birds
fig = plt.figure(figsize=(10, 10))
for i, im_name in enumerate(os.listdir(bird_dir_path)):
    full_im_path = os.path.join(bird_dir_path, im_name)
    im = load_image(full_im_path)
    norm_im = prepare_image(im, device='cpu')
    pred = predict(norm_im, vgg16, labels)
     # create a figure, just like in matlab
    ax = fig.add_subplot(1, 3 ,i+1) # create a subplot of certain size
    ax.imshow(im)
    ax.set_title(f'{im_name} is {pred}')
    ax.set_axis_off()
# plot our bird
our_bird_path = 'my_data/our_image.jpeg'
im = load_image(our_bird_path)
norm_im = prepare_image(im, device='cpu')
pred = predict(norm_im, vgg16, labels)
 # create a figure, just like in matlab
ax = fig.add_subplot(1, 3 ,2+1) # create a subplot of certain size
ax.imshow(im)
ax.set_title(f'our image is {pred}')
ax.set_axis_off()
```

```python
tr_pipeline_list = transforms.Compose([transforms.Resize(size),
                                         transforms.ToTensor(),
                                         transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                                              std=[0.229, 0.224, 0.225])])
size = 244

im = cv2.imread(our_bird_path)
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
## APPLY TRANSFORM - rot + scale
rows,cols, _ = im_rgb.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),30,1.25)
im_rgb_rot = cv2.warpAffine(im_rgb,M,(cols,rows))
## hsv color space
hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
## haussian blur
blurred = cv2.GaussianBlur(im_rgb,(13,13),cv2.BORDER_DEFAULT)

fig = plt.figure(figsize=(10, 10))
for i, (im_rgb_tr, desc) in enumerate(zip([im_rgb_rot, hsv, blurred], ['rotation+scale', 'hsv color space', 'gaussian blur'])): 
    ##FINISH TRANSFORM
    im_pil = Image.fromarray(im_rgb_tr) 
    im = tr_pipeline(im_pil)
    im = im.unsqueeze(0)
    im = torch.autograd.Variable(im)
    im = im.float()
    res = vgg16(im)
    pred = res.data.numpy().argmax()
    ax = fig.add_subplot(1, 3 ,i+1) # create a subplot of certain size
    ax.imshow(im_rgb_tr)
    ax.set_title(f'{desc} - {labels[pred]}')
    ax.set_axis_off()
```

```python
from part2.visualize_filters import plot_weights, filter_outputs
from part2.pre_process import transform_from_cv2
vgg16.eval()
# visualize filters
plot_weights(vgg16, 0, single_channel=False)
# visualize filter response

im = load_image(our_bird_path)
norm_im = prepare_image(im, device='cpu')
# regular image
reg_im_name = 'regular'
filter_outputs(norm_im, vgg16, 0, 3, reg_im_name)
# rotated image
rot_im_name = 'rotated+scaled'
rot_im = transform_from_cv2(im_rgb_rot)
filter_outputs(rot_im, vgg16, 0, 3, rot_im_name)
# different color base image
diff_col_im_name = 'hsv color space'
diff_col_im = transform_from_cv2(hsv)
filter_outputs(diff_col_im, vgg16, 0, 3, diff_col_im_name)
## haussian blur
blurred_im_name = 'blurred'
blurred_im = transform_from_cv2(blurred)
filter_outputs(blurred_im, vgg16, 0, 3, blurred_im_name)
```

```python
# get fc vector from layer number 6 and prepare training and test data
from part2.pre_process import load_image, save_feature_vec
vgg16.eval()
dirs_train = ['./dogs', './cats']
# dirs_test = ['our_dogs', 'our_cats']
# dir_to_label = {'dogs': 0,'our_dogs':0, 'cats': 1, 'our_cats': 1}
dirs_test = ['my_data/our_animals']
label_to_type = {0: 'dog', 1: 'cat'}
dir_to_label = {'./dogs': 0, './cats': 1, 'my_data/our_animals': None}
num_fc_layer = 6
save_feature_vec(dirs_train, vgg16, num_fc_layer, dir_to_label, 'train_features')
save_feature_vec(dirs_test, vgg16, num_fc_layer, dir_to_label, 'test_features')
```

```python
import numpy as np
from sklearn.svm import LinearSVC
# prepare data for training
data_train = torch.load('train_features')
X_train = np.array(data_train['X'])
y_train = np.array(data_train['y'])
clf = LinearSVC()
clf.fit(X_train, y_train)

data_test = torch.load('test_features')
X_test = np.array(data_test['X'])

y_res = clf.predict(X_test)
# print(y_res)

fig = plt.figure(figsize=(10, 10))
for i, im_name in enumerate(sorted(os.listdir('my_data/our_animals'))):
    full_im_path = os.path.join('my_data/our_animals', im_name)
    im = load_image(full_im_path)
     # create a figure, just like in matlab
    ax = fig.add_subplot(2, 2 ,i+1) # create a subplot of certain size
    ax.imshow(im)
    ax.set_title(f'{im_name}, Prediction: {label_to_type[y_res[i]]}')
    ax.set_axis_off()

    
```
